```python
import requests, re
from bs4 import BeautifulSoup

for i in range(1, 6):
    print("페이지: ", i)
    url = "https://browse.gmarket.co.kr/search?keyword=%EB%85%B8%ED%8A%B8%EB%B6%81&k=32&p={}".format(i)
    res = requests.get(url)
    res.raise_for_status()

    soup = BeautifulSoup(res.text, "lxml")
    items = soup.find_all('div', attrs={"class":re.compile("^box__component box__component-itemcard box__component-itemcard--general")})

    for item in items:
    
        name = item.find("span", attrs={"class":"text__item"}).get_text()              # 제품명

        if "Apple" in name:
            continue

        price = item.find("strong", attrs={"class":"text text__value"}).get_text()      # 가격
    
        rate = item.find("span", attrs={"class":"image__awards-points"})                # 평점
        if rate:
            rate = rate.get_text()
            rate = rate[4:6]
        else:
            continue

        link = item.find("a", attrs={"class":"link__item"})["href"]

        ad_badge = item.find("span", attrs={"class":"text"}).get_text()
        if "광고" in ad_badge:
            print("  <광고 상품 제외합니다.>")
        elif "공식" in ad_badge:
            print("  <공식 제품 제외합니다.>")
            continue

        if int(rate) >= 90:    
            print(f"제품명: {name}, 가격: {price}, 평점정보: {rate}")
            print(f"바로가기: {link}")
            print("-"*100)
```
