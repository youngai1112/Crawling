## XPath
- html은 복잡한 element로 구성된 문서
- 특정 태그의 특정 값을 불러오고 싶을 때 사용하는 경로

## 예시
- 비슷한 태그 또는 비슷한 엘리먼트가 있을 경우, 어떤 엘리먼트를 지칭하는지 명확하게 하기 위해서 XPath를 사용
- /학교/학년/반/학생[2] -> 2학년 3반의 유재석을 지칭하고 싶을 때
- /html/body/div/span/a...
- 학번이 설정되었을 경우 -> 고유한 값을 가지게 됨
    - 즉, 학번 만으로 특정 엘리먼트를 가리킬 수 있다.
    - 유니크한 값이 있을 경우, 전체 경로 대신 아래와 같이 줄여서 사용할 수 있다. 
    - //*[@학번="1-1-5"]
    - `//` : 하위 엘리먼트 전체에 해당
    - `*` :  동일
    - `@` : 속성, 한 개만 존재하는 엘리먼트일 경우
- 학교는 학년의 부모, 학년은 부모의 자식
- 학년은 반의 부모, 반은 학년의 자식
- 학생과 학생은 형제

<학교 이름 = "나도고등학교">
    <학년 value="1학년">
        <반 value="1반">
            <학생 value="1번" 학번="1-1-1">이지은</학생>
            <학생 value="2번" 학번="1-1-2">유재석</학생>
            <학생 value="1번" 학번="1-1-3">조세호</학생>
            <학생 value="1번" 학번="1-1-4">박명수</학생>
            <학생 value="1번" 학번="1-1-5">이지은</학생>
        </반>
        <반 value="2반">
        <반 value="3반">
        <반 value="4반">
    </학년>
