## 프레임워크
- 미리 만들어 놓은 블록세트
- Java로 만든 프레임워크: Spring
- Node.js로 만든 프레임워크: Express
- Python으로 만든 프레임워크: Django, Flask
- 프레임워크 안에 있는 기능 중에 내가 필요한 기능을 찾아서 사용한다. 
- 어떤 것이 더 좋은 것?
    - 좋고 나쁜 것이 아니라, 특성이 다른 것이다.
    - 언어엔 특성만 있을 뿐, 상황을 고려해 선택을 하자.
- 언어별 용처와 특징
    - 프론트엔드: 우리 눈에 보이는 부분들
        - 웹 서버는 HTML, CSS, JS 갖고 있다가 브라우저로부터 요청이 들어오면 요청 그대로 실행
        - HTML: 뼈대, CSS: 꾸미기, JS: 움직이게
        - 이 세가지만 있으면 프레임워크 없이 만들 수 있다.
        - 그러나, 위의 세가지를 더 잘 쓸 수 있게 하는 프레임 워크가 있다.
            - 리액트(React.js), 뷰(Vue.js), 앵귤러(Angular) -> 대세는 리액트
            - 자바스크립트 기반의 프레임워크
            - 용량을 더 적게 보낼 수 있다던지, 페이지의 깜빡거림이 조금 줄어드는 것 등을 할 수 있다.
    - 서버: HTML, CSS, JS를 제외한 우리가 배우는 거의 모든 언어에 해당
        - Java: 어렵다. 형식이 많은 큰 회사에서 많이 쓴다. 안정적이고, 에러 대응하기 좋다.
        - Python: 쉽게 입문. 자유로움. 데이터분석에도 사용 가능
        - Node.js: Javascript를 서버 버전으로 만든 것. 스타트업에서 많이 사용
        - (C언어) 임베디드 칩: IOT에서 사용. 가끔 게임회사를 제외하고는 C로 서버 만드는 곳은 거의 없다.
            - C언어는 컴퓨터 친화적 언어이므로, 조금 더 정교하게 명령을 내릴 수 있다.
        - (C언어) GO: 최근 각광받기 시작. 쉬운데 적당히 형식이 있음
        - (C언어) Typescript: Javascript + Type(형식) 
            - Javascript의 특성상 언어 자체가 자유롭기 때문에, 에러 발생 구역을 찾기가 어렵다.
