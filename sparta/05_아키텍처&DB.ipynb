## 서버의 실체
1. 스타트업 초창기
- 클라이언트와 서버
    - 서버 안에는 웹서버 + DB
2. 회사가 조금 더 커졌을 때
    - 유저가 많아져서 데이터 저장소를 따로 뺌
    - 클라이언트 - 서버 - DB (웹 서버와 DB 분리)
3. 더 커졌을 때
    - 클라이언트 - 라우터 (교통순경) - 웹서버 여러개 - DB
        - 라우터의 역할 : 요청이 들어오면, 비어있는 웹 서버에 빨리 요청을 줌
        - DB : 데이터는 한 곳에서 관리를 하는 게 거의 통상적
    - 이미지가 있을 경우 (이미지 따로 저장)
        - 클라이언트 - 라우터 (교통순경) - 웹서버 여러개 - 이미지용 DB / 텍스트용 DB
4. 더더 커졌을 때
    - 서버상태 체크 (갑자기 서버 터지는 것 방지)
    - 클라이언트 - 라우터 (교통순경) - 웹서버 여러개 - 이미지용 DB / 텍스트용 DB - 상태체크용 서버 - 팀메신저/전화/상황판

## DB (데이터베이스)
- 잘 보관하기 위해? 잘 찾기 위해?
- 데이터베이스 : 데이터를 잘 찾게 해주는 프로그램
- 예시 : 경영 섹션 > 출판사 별 > 책 제목 별로 정리되어 있을 때, 뭐가 쉬울까?
    - 1) '사랑'이 제목에 들어간 책 다 주세요 ---> 인덱스를 타지 않는 경우 [풀 스캐닝]
    - 2) '한국출판사' 책 다 주세요  ------------> 인덱스를 타는 경우 / 더 빠르게 데이터를 찾을 수 있다.
- DB도 정렬된 상태대로 찾아야 빠르게 찾아온다.
- 인덱스 : 정리된 항목, 인덱스틑 타야 빠르게 가져올 수 있다.
- DB의 종류 
    - SQL : 정형화, 표를 먼저 그리고 데이터를 넣는다. 안정성
        - ORACLE, MySQL, ...
    - NoSQL(Not-only SQL) : 비정형화, JSON 형식으로 데이터를 쌓을 수 있다.
        - mongoDB, ...
- 데이터분석 하기위해 SQL을 배운다 = SQL DB에서 데이터를 빼내는 방법(language)을 배운다.
