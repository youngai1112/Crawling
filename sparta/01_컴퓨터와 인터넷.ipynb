## 컴퓨터
- 기본적으로 기계를 모아둔 것
- 하드디스크 : 적어두기, 하드디스크가 크면 여러 가지 용량을 저장해 둘 수 있다.
- RAM : 기억하기, PAM이 좋으면 기억을 많이할 수 있다.
- CPU : 계산하기, CPU가 좋으면 계산을 빨리할 수 있다.
- 메인보드 : 꽂아두기

## 컴퓨터를 조작하는 방법
- 운영체제라는 것을 통해서 조작
- 운영체제 (OS) : 인간이 기계와 소통할 수 있게 만드는 기반 프로그램
- 예를 들어, 안드로이드, IOS, Windows10, LINUX(오픈 소스)가 있다.

## 코딩
- OS에 명령을 내리는 게 코드
    - OS에 명령을 내려서 기계를 제어

## 프로그래밍 언어
- 영어로 할거냐, 이탈리아어로 할거냐의 차이
    - 즉, 어떤 말로든 명령만 내리면 됨 
    - 각 언어마다 특성이 존재함

## 프로그램
- 프로그래밍으로 코딩을 해서 매번 똑같이 움직이는 것
- 클릭하면, ppt 파일을 읽어서 보여달라 요청
- 코딩은 늘 프로그램을 만드는 것

## 인터넷
- 컴퓨터끼리 대화를 주고 받고 싶을 때 사용하는 것이 인터넷

## 웹
- 컴퓨터끼리 대화를 주고 받을 때의 규칙이 필요
    - 일단 이름 말하고, 무슨 프로그램 찾는지 말해 그 다음에는...
- 그러한 규칙 중 하나가 웹 : http
- `http` : 웹 통신을 하겠다는 의미 
    - "내가 지금부터 웹 규칙대로 말할 거니까 잘 들어"
- `https` : 보안을 강화
    - 암구호 표를 서로 가져서 가운데서 못 듣게 하는 역할

## 웹 서버
- "잘 듣고 있다가 누가 http 요청하면 이 파일을 돌려줘"
