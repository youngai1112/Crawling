## 서버
- 우리가 사람들이 같이 쓰는 서비스를 만들려고 함 > 데이터 관리 역할 필요 (데이터가 일관성있게 나가기 위함)
- 가운데서 데이터를 저장하고, 주는 것
- 그러니까 사람이 많아지면 힘들다!
- 서버가 터지면 카톡이 안되는 이유
    - 서버는 전달자 역할을 한다. 카톡 서버가 편지 전달을 못하게 되어서 그렇다.
- "잘 듣고 있다가 누가 http 요청하면 이 파일을 돌려줘" == 웹 서버를 만들었다고 하는 것이다.
- 서버란, '컴퓨터'라는 사람의 직업과 같은 것이다.
    - 즉, 우리가 쓰는 프로그램 중 하나이다.

## 브라우저의 역할
- www.naver.com을 치면 서버에 요청을 보낸다. > 네이버 서버 컴퓨터에서 뭔가를 내려준다.
    - 네이버 서버 컴퓨터에서 내려준 것을 그대로 그려주는 것이 브라우저의 역할
    - `뼈대`와 `꾸미기`, `움직이기`
    - HTML, CSS, JavaScript를 보내주고, 브라우저는 그대로 그려주기만 한다.
- 요약 : 요청을 보내고 받아서 그대로 그려주는 것

## 페이지를 주는 또다른 방법
- 콘서트 티켓팅 화면처럼 데이터만 갈아끼우고 싶을 수도 있다.
    - 데이터만 내려주는 것도 가능하다. (CSS, JavaScript는 안 내려받으면 됨)
- 데이터만 내려줄 때의 규칙 : JSON 형식
    - {Key : Value}
    - 중괄호로 이루어져 있는 것이 JSON의 특징
- JSON : 통신할 때, 데이터를 받아올 때 JSON 방식으로 받아온다.
    - 앱에서 자주 사용 : 껍데기만 가지고 있다가, 데이터를 받아서 붙여주는 것
        - 그래서, 업데이트를 안하면 카톡에 눈이 오지 않는 것!
    - 웹에서는 : 서버에서 껍데기까지 다 받아온다.
        - 그래서, 업데이트를 안했기 때문에 쿠팡 크리스마스 이벤트가 안보인다? X
