## 서버-클라이언트 통신
- 은행을 서버라고 한다면
    - 예를 들어, 은행에는 수많은 데이터가 있으니까 은행을 서버라고 가정. 고객은 클라이언트 (요청하는 쪽)
    - 고객의 요청 "잔고를 확인하고 싶어요"
    - 은행과 미리 약속해둔 곳에 [예금창구], 약속해둔 데이터를 들고가야 [신분증] 잔고를 알려준다. [응답]
- 컴퓨터로 돌아와서 생각해보자
    - 예시) 배달의 민족
    - A: 아이디를 갖고 와. 주문내역을 알려줄게
    - B: 주소를 갖고 와. 음식점을 알려줄게
    - C: 주문번호를 갖고 와. 배달현황을 알려줄게
- `API` : Application Programming Interface
    - 서버에서 미리 만들어둔 기능이자 약속
        - 클라이언트 개발자와 서버 개발자가 미리 한 약속
        - 예를 들어, "음식점 리스트를 부르실 때는 B로 요청 주시구요. 주소는 꼭 가져와 주세요.그러면 제가 이렇게 정보를 드릴게요."
    - API 명세서 : 요청할 것들이 많아져서, 문서로 써둔 것
        - Request : 이렇게 요청하시면
        - Response : 제가 드리겠습니다.
    - 어떻게 생겼냐면
        - 우리은행 용산지점의 입출금창구에 가는데 주민등록번호를 가지고 가야 한다.
            - http://`우리은행 용산지점`/`입출금창구`?신분증=`주민등록번호`
            - `우리은행 용산지점` : 서버위치, `입출금창구`: API주소, `주민등록번호`: 가져가기로 한 데이터
- 회사에서 쓰이는 예시
    - "신규피쳐라 API 하나만 따주세요" = 새로운 기능 창구를 하나 만들어달라. 그쪽으로 요청하겠다.
    - "API 개발은 끝났고, 클라이언트는 아직입니다" = 창구는 다 만들었는데, 요청하는 쪽이 아직 작업중
- `OPEN API`
    - 누구나 요청할 수 있게 열어둔 API
        - 통상적으로는 누구나 부르지 못하게 한다.

- 정리
    - 클라이언트 : 요청을 하는 쪽 ex. 브라우저(HTML,CSS,JS), 휴대폰(JSON), IOT가전(JSON)
    - 서버 : 요청을 받고, 데이터를 주는 쪽
    - API : 요청을 받을 때의 출입문
    - 웹 서버 : 요청을 잘 받기 위해서 대기하고 있다가, API를 통해 요청이 들어오면 수행한다.
        - 요청에 따라 자신이 가지고 있던 HTML, CSS, JS를 클라이언트(브라우저)쪽에 줄 수 있다.
        - 데이터를 JSON 형식으로 앱에 내려줄 수도 있음.
