`goodscode` 찾기
- href에 들어있는 `goodscode`를 찾아보자
- `attribute`에 있는 값을 가져오기 위해서는 `get(attribute name)`

```python
link_item1 = item_best1.select_one("a.itemname").get("href")
print(link_item1)
```

*url의 ? 뒤부터 parameter <= client가 server에게 보내는 값*

*? 뒤에 있는 파라미터 나열은 쿼리 스트링이라 한다*
* goodscode와 ver이라는 정보를 보냄
* 우리에게 필요한 것은 goodscode

```python
link_goodscode = link_item1.split("/")[3]
link_goodscode = link_goodscode.split("=")[1]
link_goodscode = link_goodscode.split("&")[0]
link_goodscode
```

```python
# 1. split("?") 
# 2. [1]
# 3. split("&")
# 4. [0]
```

`urllib` 라이브러리
- 기본적인 웹 통신 및 url을 관리하기 위한 여러 기능들이 들어있는 모듈
- `urlparse`, `parse_qs`

```python
from urllib.parse import urlparse, parse_qs
```

```python
# urlparse
parsed_url = urlparse(link_item1)
parsed_url
```

```python
url_query = parsed_url.query
url_query
```

```python
# parsed_qs url의 쿼리 부분을 딕셔너리 형태로 바꿔준다.
parsed_query = parse_qs(url_query)
parsed_query
```

```python
item1_goodscode = parsed_query["goodscode"][0]
item1_goodscode
```

`image` url 가져오기

```python
img_url = item_best1.select_one("div.thumb > a > img").get("data-original") # 이미지는 어트리뷰트 값으로 들어옴
img_url
```

`img` 저장 라이브러리 - `urllib.request.urlretrieve`

```python
from urllib.request import urlretrieve

# urlretrieve(이미지 링크, 저장할 파일명)
# 임의로 png라고 적어준다. 이미지 확장자가 있으면 그냥 그거 그대로 쓰는게 좋음
urlretrieve(img_url, item1_goodscode+".png") 
# 지마켓의 경우, 확장자 자체가 없다
# 보통은 img src="dddd/tt/이미지이름.jpg" => 확장자가 있는 경우 jpg를 쓰자
```
